```python
import pandas 

df = pandas.read_csv( 'static/neat_plain_text_template.csv', index_col=0 )

df
```

```python
from itertools import product

rows = [ 'A', 'C', 'E', 'G', 'I', 'K' ]
cols = [ '01', '03', '05', '07', '09', '11' ] 
my_wells = [''.join(n) for n in product( rows, cols )]
```

```python
for w in my_wells:
    df.loc[w] = ('sample_' + w,'t7','t7term')
    
df.to_csv( 'static/sparse_example_order.csv' )
```
