```python
from sqlalchemy import create_engine
from io import StringIO
import pandas 

engine = create_engine( 'sqlite:////tmp/neat.db' ) 
conn = engine.connect()
```

```python
transfers = pandas.DataFrame( None, columns=['source', 'dest', 'vol'] )
sample = 0 

r = conn.execute( 'select * from sequencing' )
print( 'Load plate', r.first() ) 
for seq_id, cust_id, d in r:
    df = pandas.read_csv( StringIO( d ), index_col=0 )
    for idx, row in df.iterrows():
        # all primers 100 µM 
        # PCR setup: 
        # 5 nL primer 1
        # 5 nL primer 2
        # 5 nL template 
        n = ( idx, sample % 1536, '5 nL' )
        sample += 1 
        print( 'Xfer {2} from {0} to {1} on plate {3}'.format( *n, sample // 1536 ) )
        
```
